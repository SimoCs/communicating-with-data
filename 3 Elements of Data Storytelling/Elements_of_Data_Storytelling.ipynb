## Examples of Good vs Bad Data Storytelling

### Setup:

```python
import matplotlib.pyplot as plt

import numpy as np


# Sample data

years = np.array([2015, 2016, 2017, 2018, 2019, 2020])

sales = np.array([50, 55, 52, 58, 63, 70])

expenses = np.array([48, 50, 49, 53, 60, 68])
```

### Good Data Storytelling:
*   A clear title
*   Labelled axes
*   A legend
*   A color scheme that is easy on the eyes
*   Annotations to highlight specific points of interest

```python
plt.figure(figsize=(14,7))

plt.plot(years, sales, marker='o', label='Sales', color='blue')

plt.plot(years, expenses, marker='o', label='Expenses', color='red')

plt.title('Yearly Sales vs Expenses (2015-2020)', fontsize=15)

plt.xlabel('Year', fontsize=12)

plt.ylabel('Amount ($)', fontsize=12)

plt.legend()

plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.annotate('Highest Growth', xy=(2019, 63), xytext=(2016,62),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=9)

plt.tight_layout()

plt.show()
```

### Bad Data Storytelling:
*   No title or unclear title
*   No axis labels
*   Clashing or too many colors
*   Overloaded with information or unnecessary embellishments
*   No legend or unclear legend

```python
plt.figure(figsize=(10,6))

plt.plot(years, sales, marker='o', color='green')

plt.plot(years, expenses, marker='o', color='yellow')

plt.xlabel('Time Period')

plt.ylabel('Amount')

plt.grid(True, which='both', linestyle='-', linewidth=2)

plt.show()
```

**In the bad example:**
*   The colors used might not differentiate the lines well, especially for those with color vision deficiencies.
*   The grid lines are too bold and distracting.
*   It's unclear what the two lines represent without a legend or clear labels.
